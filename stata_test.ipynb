```python
import pandas as pd
import matplotlib.pyplot as pyplot
url = "https://www.dropbox.com/s/ie1zvkr7iht2i1j/public_workingtimedata.dta?dl=1"

df = pd.read_stata(url)


df.to_csv('./csv.csv')

df_by_year = df.sort_values(["year"])
df_US = df.loc[df['country_name'] == "United States of America"]

df_US_hours = df_US[["country_name", "year", "normalhours"]]

df_US_hours.plot.scatter(x="year", y="normalhours")


df_2020 = df.loc[df["year"] == 2020]
df_2020.plot.scatter(x="normalhours", y="hours_max")
```
